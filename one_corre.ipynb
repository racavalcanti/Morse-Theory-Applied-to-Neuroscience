```python
def split_into_25(ind):
    list_ind = li[ind]
    data_with_25 = []
    data_array = list_ind.to_numpy()
    num_rows = data_array.shape[0]
    window_size = 25
    range_2 = range(0, num_rows - window_size + 1)
    for n in range_2:
        data_slice = data_array[n:n + window_size, :]
        dfr = pd.DataFrame(data_slice, columns=list_ind.columns)
        data_with_25.append(dfr)
        
    return data_with_25

#116 é o ideal para spearman
```

```python
all_list_with_25_cols = []
for m in range(len(li)):
    all_list_with_25_cols.append(split_into_25(m))
```

```python
def fract_euler(ind):
    all_list_with_25_cols_ind = all_list_with_25_cols[ind]
    correlation_matrices_ind = []
    for n in range(len(all_list_with_25_cols_ind)):
        matrix_corr = all_list_with_25_cols_ind[n].corr(method = "spearman")
        matrix_corr = np.array(matrix_corr)
        correlation_matrices_ind.append(matrix_corr)

    eigens_ind = []
    for m in range(len(correlation_matrices_ind)):
        corre_m = correlation_matrices_ind[m]
        w,v = np.linalg.eigh(corre_m)
        eigens_ind.append(w)

    list_negative_eigenss_ind = [[value for value in sublist if value < 0] for sublist in eigens_ind]
    
    list_negative_eigenss_tresh_ind = [[value for value in sublist if value < -0.00000000000001] for sublist in eigens_ind]

    qtd_negative_eigens_ind = []
    for n in range(len(list_negative_eigenss_ind)):
        list_negative_eigenss_ind_n = list_negative_eigenss_ind[n]
        qtd_negative_eigens_ind.append(len(list_negative_eigenss_ind_n))
        
    
    qtd_negative_eigens_tresh_ind = []
    for n in range(len(list_negative_eigenss_tresh_ind)):
        list_negative_eigenss_tresh_ind_n = list_negative_eigenss_tresh_ind[n]
        qtd_negative_eigens_tresh_ind.append(len(list_negative_eigenss_tresh_ind_n))


    signal_euler_charac_ind = []
    for n in range(len(qtd_negative_eigens_ind)):
        qtd_negative_eigens_ind_n = qtd_negative_eigens_ind[n]
        a = (-1)**qtd_negative_eigens_ind_n
        signal_euler_charac_ind.append(a)
        
    signal_euler_charac_tresh_ind = []
    for n in range(len(qtd_negative_eigens_tresh_ind)):
        qtd_negative_eigens_tresh_ind_n = qtd_negative_eigens_tresh_ind[n]
        a = (-1)**qtd_negative_eigens_tresh_ind_n
        signal_euler_charac_tresh_ind.append(a)
    
    
    skewness_no_moment = skew(signal_euler_charac_ind)
    skewness_moment_tresh = skew(signal_euler_charac_tresh_ind)
    kurtosis_moment = kurtosis(signal_euler_charac_ind)
    kurtosis_moment_tresh = kurtosis(signal_euler_charac_tresh_ind)
    
    
    return {'skewness_no_moment': skewness_no_moment, 'skewness_moment_tresh':skewness_moment_tresh, 'kurtosis_moment':kurtosis_moment, 'kurtosis_moment_tresh': kurtosis_moment_tresh }
```

```python
skewness_no_tresh = []
skewness_with_tresh = []
kurtosis_no_tresh = []
kurtosis_with_tresh = []

for i in range(0,len(names)):
    skewness_no_tresh.append(fract_euler(i)['skewness_no_moment'])
    skewness_with_tresh.append(fract_euler(i)['skewness_moment_tresh'])
    kurtosis_no_tresh.append(fract_euler(i)['kurtosis_moment'])
    kurtosis_with_tresh.append(fract_euler(i)['kurtosis_moment_tresh'])
#salve em txt
```
